```python
#Pudimos comprobar que todos los datos tienen el mismo CRS, entonces procedemos con el siguiente paso.
#Comenzamos creando la base, en este caso será "estados" y luego agregamos las otras dos capas en 
#la parte superior 
base = estados.plot(facecolor="white", edgecolor='black', linewidth=0.1,figsize=(12,12))
cities.plot(marker='.', color='red', markersize=1,alpha=0.7,
            ax=base) # on top of..
rivers.plot(edgecolor='blue', linewidth=0.5,
            ax=base)# on top of..
```

```python
#En esta parte importamos folium, que es una libreria que nos ayudará a representar el mapa de 
#una forma interectiva
import folium


m = cities.explore(color="red", 
                   name="cities")#optional

m = rivers.explore(m=m, color="blue",
                   name="rivers")#optional
folium.LayerControl().add_to(m) #optional
m
```

```python
#Creamos un subconjunto del mapa de Mexico y nos centramos en el estado de Nayarit
nayarit=estados[estados.ESTADO=='Nayarit']
nayarit
```

```python
#Nos quedamos con las partes de los otros mapas que se cruzan con nuestro subpoligono.
cities_clipped = gpd.clip(gdf=cities,
                          mask=nayarit)
rivers_clipped = gpd.clip(gdf=rivers,
                               mask=nayarit)
```

```python
#De la misma forma, trazamos la version recortada del estado de Nayarit
base = nayarit.plot(facecolor="greenyellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
cities_clipped.plot(marker='+', color='red', markersize=8,ax=base)
rivers_clipped.plot(edgecolor='blue', linewidth=0.5,ax=base)
```

```python
#Buscamos las coordenadas que nos aproximadamente el centro de Nayarit 
nayaritCoord = [21.50951, -104.89569]
```

```python
#Gracias a las coordenadas que conseguimos, podemos hacer el mapa de manera interactiva y podemos
#centrarlo en la coordenada que queriamos. 
m = cities.explore(location=nayaritCoord,
                   zoom_start=7.5,
                   tiles='CartoDB positron',
                   color='red',
                   name="cities") #optional
m = rivers.explore(m=m, color="blue",
                   name="rivers")#optional
# folium.LayerControl().add_to(m) #optional
m
```

```python
#Guardamos el mapa como imagen 
import matplotlib.pyplot as plt

base = nayarit.plot(facecolor="greenyellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
map2=cities_clipped.plot(marker='+', color='red', markersize=15,ax=base)
mapEnd=rivers_clipped.plot(edgecolor='blue', linewidth=0.5,ax=base)
plt.savefig(os.path.join("maps",'mapEnd.jpg'))
```
